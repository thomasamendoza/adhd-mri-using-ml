```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import nibabel as nib # common way of importing nibabel

mri_file = 'anat.nii.gz'
img = nib.load(mri_file)

# print(type(img))
print("dimensions in voxels", img.shape)
print("voxel size", img.header.get_zooms())
print("units of voxels and time", img.header.get_xyzt_units())


img_data = img.get_fdata()
print("data type of operable data", type(img_data))
print(img_data.shape)
mid_slice_x = img_data[120, :, :]
print(mid_slice_x.shape)

plt.imshow(mid_slice_x.T, cmap='gray', origin='lower')
plt.xlabel('First axis')
plt.ylabel('Second axis')
plt.colorbar(label='Signal intensity')
plt.show()
```
